```python
import pandas as pd

# cp949 인코딩으로 CSV 파일 읽기
df = pd.read_csv('JEJU_MCT_DATA.csv', encoding='cp949')

# 데이터프레임의 처음 몇 행 출력하여 확인
print(df.head())
```

```python

# MCT_NM과 OP_YMD를 결합한 새로운 컬럼 생성
df['MCT_YM'] = df['MCT_NM'] + '_' + df['OP_YMD'].astype(str)

# MCT_YM 그룹의 개수 계산
group_counts = df['MCT_YM'].value_counts()

# 12개 이상의 데이터를 가진 그룹만 유지
df = df[df['MCT_YM'].isin(group_counts[group_counts == 12].index)]
```

```python

# 각 열의 첫 글자만 추출하는 함수를 정의합니다
def extract_first_char(value):
    return value.split('_')[0]

# 처리할 열 목록
columns_to_process = ['UE_CNT_GRP', 'UE_AMT_GRP', 'UE_AMT_PER_TRSN_GRP']

# 각 열에 대해 처리를 수행합니다
for column in columns_to_process:
    # 새로운 열 이름을 생성합니다
    new_column = f"{column}_NUM"
    
    # 새로운 열을 생성하고 첫 글자만 추출하여 저장합니다
    df[new_column] = df[column].apply(extract_first_char)
    
    # 결과를 출력합니다
    print(f"\n{new_column}의 고유 값:")
    print(df[new_column].unique())
    print(f"고유 값의 개수: {df[new_column].nunique()}")

# 변경된 데이터프레임의 처음 몇 행을 출력합니다
print("\n변경된 데이터프레임의 처음 5행:")
print(df[columns_to_process + [col + '_NUM' for col in columns_to_process]].head())
```
