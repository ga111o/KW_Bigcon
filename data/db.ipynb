```python
import pandas as pd
import sqlite3


csv = pd.read_csv("./FDA.csv")

id_pk = csv['key'].tolist()
place_ids = csv['placeID'].tolist()
MCT_NMS = csv['MCT_NM'].tolist()


data = pd.DataFrame({'id': id_pk, 'placeID': place_ids, 'MCT_NM': MCT_NMS})

data = data.drop_duplicates()

data = data.sort_values(by='placeID')

conn = sqlite3.connect('jeju2.db')
cursor = conn.cursor()

for index, row in data.iterrows():
    try:
        print(f"{row['id']}\t{row['placeID']}\t{row['MCT_NM']}")
        cursor.execute(
            'INSERT INTO Information (id, placeID, MCT_NM) VALUES (?, ?, ?)', (row['id'], row['placeID'], row['MCT_NM']))
    except sqlite3.IntegrityError:

        continue


conn.commit()
cursor.close()
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect('./jeju2.db')
cursor = conn.cursor()

cursor.execute("SELECT id, placeID FROM Information;")
rows = cursor.fetchall()

for row in rows:
    id = row[0]
    placeID = row[1]
    if placeID == None:
        continue

    if '.' in placeID:

        new_placeID = placeID.split('.')[0]

        cursor.execute(
            "UPDATE Information SET placeID = ? WHERE id = ?", (new_placeID, id))

conn.commit()

cursor.close()
conn.close()
```

```python
import sqlite3


conn = sqlite3.connect('./place.db')
cursor = conn.cursor()

cursor.execute(
    "SELECT key, placeID FROM restaurant WHERE key IS NOT NULL AND placeID IS NOT NULL;")

rows = cursor.fetchall()

conn22 = sqlite3.connect('jeju2.db')
cursor22 = conn22.cursor()

for row in rows:

    print(f"key {row[0]}, placeID {row[1]}")
    cursor22.execute(
        'UPDATE Information set placeID = ? where id = ?', (row[1], row[0]))

conn22.commit()

cursor.close()
conn.close()
conn22.close()
```
