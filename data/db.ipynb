```python
import pandas as pd
import sqlite3


csv = pd.read_csv("./crawling.csv", encoding="cp949")


place_ids = csv['placeID'].tolist()
MCT_NMS = csv['MCT_NM'].tolist()


data = pd.DataFrame({'placeID': place_ids, 'MCT_NM': MCT_NMS})


data = data.drop_duplicates()


data = data.sort_values(by='placeID')


conn = sqlite3.connect('jeju.db')
cursor = conn.cursor()


for index, row in data.iterrows():
    try:
        cursor.execute(
            'INSERT INTO Information (placeID, MCT_NM) VALUES (?, ?)', (row['placeID'], row['MCT_NM']))
    except sqlite3.IntegrityError:

        continue


conn.commit()
cursor.close()
conn.close()
```

```python
import pandas as pd
import sqlite3

#
csv = pd.read_csv("./crawling.csv", encoding="cp949")

place_ids = csv['placeID'].tolist()
ADDRs = csv['ADDR'].tolist()
MCT_TYPEs = csv['MCT_TYPE'].tolist()


conn = sqlite3.connect('jeju.db')
cursor = conn.cursor()


for row in zip(ADDRs, MCT_TYPEs, place_ids):

    cursor.execute(
        'UPDATE Information SET ADDR = ?, MCT_TYPE = ? WHERE placeID = ?', row)


conn.commit()

cursor.close()
conn.close()
```
